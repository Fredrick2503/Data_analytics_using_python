```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("btp-station-wise-accidents-2023.csv")
data
```

```python
data.rename(columns={"2023 - Total Cases":"Total cases"},inplace=True)
data.drop(['Sub-division',"2023 - Fatal Cases","2023 - Killed People","2023 - Non-Fatal","2023 - Injured People"],axis=1,inplace=True)
totals_data=data[data["Station"].str.contains("Total")]
data=data[(data["Station"].str.contains("Total"))!=True]
```

```python
print("Data after cleaning the data")
display(data)
```

```python
Zones=data["Zone"].drop_duplicates()
for zone in Zones:
    plt.figure(figsize=(16,9))
    ax=sns.barplot(data=data[data['Zone']==zone],x="Station",y="Total cases",width=0.3,palette="bright")
    ax.tick_params(axis='x',rotation=90)
    ax.tick_params(axis='both',labelsize=16)
    ax.xaxis.label.set_size(20)
    ax.yaxis.label.set_size(20)
    plt.axhline(y=(data[data['Zone']==zone])['Total cases'].mean(), color='black', linestyle='--', label='Average cases')
    plt.title(f"Bar graph visualising total cases in every station in {zone} zone",size=24)
    plt.show()
    display(data[data['Zone']==zone])
```

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as mpatches
import seaborn as sns

# Data for the pie chart
Zones = data["Zone"].drop_duplicates()

for zone in Zones:
    plt.figure(figsize=(16, 9))
    zone_data = data[data['Zone'] == zone]
    stations = zone_data['Station']
    total_cases = zone_data['Total cases']
    
    # Generate a list of unique colors using the "bright" palette
    colors = sns.color_palette('bright', len(stations))
    
    # Plotting the pie chart
    wedges, texts = plt.pie(total_cases, colors=colors, startangle=140)
    
    # Adding a legend with station names and percentages
    legend_labels = [f'{station}: {percentage:.1f}%' for station, percentage in zip(stations, 100 * total_cases / total_cases.sum())]
    patches = [mpatches.Patch(facecolor=colors[i], label=legend_labels[i]) for i in range(len(stations))]
    plt.legend(handles=patches, title="Stations", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1), fontsize=16, title_fontsize=20)
    plt.title(f'Pie Chart of Total Cases in Every Station in {zone} Zone', size=24)
    plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    
    # Increase label size
    plt.setp(texts, size=16)
    
    plt.show()
    display(zone_data)
```
